```python
customers = {'CustomerID': [10, 11],
             'Name': ['MikeWang', 'JackMa'],
             'Address': ['Address for MikeWang', 'Address for JackMa']}
customers = pd.DataFrame(customers)
customers
```

```python
orders = {'CustomerID': [10, 11, 10],
          'OrderDate': [datetime.date(2016, 12, 1),
                        datetime.date(2016, 12, 1),
                        datetime.date(2016, 12, 2)]}
orders = pd.DataFrame(orders)
orders
```

```python
pd.merge(customers, orders)
```

```python
left_data = {'key1': ['a', 'b', 'c'],
             'key2': ['x', 'y', 'z'],
             'lval1': [0, 1, 2]}
right_data = {'key1': ['a', 'b', 'c'],
              'key2': ['x', 'a', 'z'],
              'lval2': [6, 7, 8]}

left_data = pd.DataFrame(left_data, index=[0, 1, 2])
right_data = pd.DataFrame(right_data, index=[0, 1, 2])
```

```python
left_data
```

```python
right_data
```

```python
pd.merge(left_data, right_data, on='key1')
```

```python
pd.merge(left_data, right_data, how='outer')
```
