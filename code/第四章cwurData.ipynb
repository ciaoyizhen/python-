```python
import pandas as pd
pd.set_option('display.max_rows', 6)
pd.set_option('display.max_columns', 6)
```

```python
file = r'../data/cwurData.csv'
cwur = pd.read_csv(file)
cwur.head()
```

```python
cwur_85 = cwur['score'].gt(85)
cwur_85
```

```python
# 选择score>85的数据
cwur_score_85 = cwur[cwur['score']>85]
cwur_score_85.head()
```

```python
# 给出85分数据以上的学校有多少,占多少比例
cwur_85 = cwur['score']>85
print(f'85分以上的学校有{cwur_85.sum()}个')
print(f'85分以上的学校占{cwur_85.mean()*100}%')
```
