```python
# 对缺失值使用均值填充
retail_data['UnitPrice'] = retail_data['UnitPrice'].\
                            fillna(retail_data['UnitPrice'].mean())
# 对缺失值使用众数填充
retail_data['Quantity'] = retail_data['Quantity'].\
                            fillna(retail_data['Quantity'].mode())
retail_data.head()
```
