```python
import pandas as pd
import numpy as np
```

```python
file = r'../data/billboard.csv'
df = pd.read_csv(file)
df.head()
```

```python
id_vars = ['year', 'artist', 'track', 'time', 'date.entered']
var_name = 'week'
value_name = 'rank'
df = pd.melt(df, id_vars=id_vars, var_name=var_name, value_name=value_name)
df.head()
```

```python
df['week'] = df['week'].str.extract('(\d+)', expand=False).astype(int)
df
```

```python
df = df.dropna()
df
```

```python
df['date'] = pd.to_datetime(df['date.entered']) + pd.to_timedelta(df['week'], unit='w') - \
                pd.DateOffset(weeks=1)  # -Offset是为了+week=1的时候减去相应的值
df.head()
```

```python
df = df.sort_values(ascending=True, by=['year', 'artist', 'track', 'time', 'week', 'rank'])
df
```

```python
songs_cols = ['year', 'artist', 'track', 'time']
songs = df[songs_cols].drop_duplicates()
songs.head()
```

```python
songs.reset_index(drop=False)
songs['song_id'] = songs.index
songs
```

```python
df
```

```python
rank = pd.merge(df, songs, on=['year', 'artist', 'track', 'time'])
rank
```
