```python
import numpy as np
import pandas as pd
```

```python
file = r'../data/tb.csv'
df = pd.read_csv(file)
df
```

```python
id_vars = ['country', 'year']
tmp_df = pd.melt(df, id_vars=id_vars, value_name='case', var_name='sex_and_age')
tmp_df.tail(20)
```

```python
new_tmp_df = tmp_df['sex_and_age'].str.extract('(\D)(\d+)(\d{2})', expand=False)
new_tmp_df.tail(20)
```

```python
new_tmp_df.columns = ['sex', 'age_lower', 'age_upper']
new_tmp_df.head()
```

```python
df = pd.concat([tmp_df, new_tmp_df], axis=1)
df.head()
```

```python
df['age'] = df['age_lower']+'-'+df['age_upper']
df.head()
```

```python
df.drop(['sex_and_age', 'age_lower', 'age_upper'], axis=1, inplace=True)
df.head()
```

```python
df.sort_values(ascending=True, by=['country', 'year', 'sex', 'age', 'case'])
```

```python
file = r'../data/weather.csv'
df = pd.read_csv(file)
df.head()
```

```python
df = pd.melt(df, id_vars=['id', 'year', 'month', 'element'], var_name='raw_day')
df.head()
```

```python
df['day'] = df['raw_day'].str.extract('d(\d+)', expand=False)
```

```python
df[['year', 'month', 'day']] = df[['year', 'month', 'day']].apply(lambda x: pd.to_numeric(x, errors='ignore'))
df
```

```python
df['day'] = df['raw_day'].str.extract('(\d+)', expand=False).astype(int)
df
```

```python
df.nlargest(10, 'day')
```

```python
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
```

```python
import datetime
```

```python
def create_date(row):
    result = datetime.datetime(year=row['year'],
                               month=row['month'],
                               day=row['day'])

df['date'] = df.apply(lambda row: create_date(row), axis=1)
df
```
