```python
def max_deviation(s):
    # s为处理列
    std_score = (s - s.mean()) / s.std()
    return std_score.abs().max()


df_tips.groupby('sex')['tip'].agg(max_deviation)
```

```python
df_tips.groupby('sex').agg({'tip': max_deviation})
```

```python
def bill_between(s, low, high):
    return s.between(low, high).mean()


df_tips.groupby('sex')['total_bill'].agg(bill_between, 30, 60)
```

```python
df_tips.groupby('day').filter(lambda x: x['total_bill'].mean() > 20)
```
