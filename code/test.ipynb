```python
import pandas as pd
import numpy as np
```

```python
boolean=[True,False]
gender=["男","女"]
color=["white","black","yellow"]
data=pd.DataFrame({
    "height":np.random.randint(150,190,100),
    "weight":np.random.randint(40,90,100),
    "smoker":[boolean[x] for x in np.random.randint(0,2,100)],
    "gender":[gender[x] for x in np.random.randint(0,2,100)],
    "age":np.random.randint(15,90,100),
    "color":[color[x] for x in np.random.randint(0,len(color),100) ]
}
)
data
```

```python
data['gender'] = data['gender'].map({'男':1, '女':0})
data
```

```python
def gender_map(x):
    return 1 if x=='男' else 0

data['gender'] = data['gender'].apply(gender_map)
```

```python
import seaborn as sns
import pandas as pd

df_tips = sns.load_dataset('tips')
df_tips

groups = df_tips.groupby(by='sex')

for title, group in groups:
    print(title)
    print(group)
    
# 可以通过get_group()函数直接获取指定组
groups.get_group('Female')
```

```python
df_tips.groupby('sex').mean()
df_tips.groupby('sex').agg('mean')
# 上述两种方法的结果是一致的。

# agg可以同时计算多个
df_tips.groupby('sex').agg(['mean', 'min', 'max'])

# agg可以更改列名
df_tips.groupby('sex')['tip'].agg([('tip_mean', 'mean'), ('tip_min', 'min'), ('tip_max', 'max')])
# 前面是列名,后面是方法

# df_tips.groupby(['sex']).agg({'tip': ('tip.mean', 'mean'),
#                              {'处理的列名': ('处理完后的名字', '处理的办法')}})  # 这里以元组的方式
```

```python
df_tips.groupby(['sex']).agg({'tip': ('tip.mean', 'mean')})
                             # {'处理的列名': ('处理完后的名字', '处理的办法')}})  # 这里以元组的方式
```
