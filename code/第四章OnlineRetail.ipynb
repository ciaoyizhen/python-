```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
```

```python
file = r'../data/OnlineRetail.csv'
retail_data = pd.read_csv(file, encoding='unicode_escape')
retail_data.head()
```

```python
retail_data['UnitPrice'] = retail_data['UnitPrice'].\
                        fillna(retail_data['UnitPrice'].mean())

retail_data['Quantity'] = retail_data['Quantity'].\
                        fillna(retail_data['Quantity'].mode())
retail_data['Total_price'] = retail_data['UnitPrice'] * retail_data['Quantity']

retail_data.head()
```
